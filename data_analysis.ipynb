2. **HERE RECURSIVE FUNCTION USE TO REDUCE THE TIME COMPLEXITY**



```python
import numbers


# normalize data because some time in particuler column defind by numpy like np.int64(4) or np.float64(4.59)
# that's why all numbers convert in normal numeric form
def normalize_value(val):
    if isinstance(val, numbers.Number):
        return val.item() if hasattr(val, 'item') else val
    return val


def for_clean(row,length,check,unique_row):

   if len(row) == length:
        row_cleaned = tuple(normalize_value(cell) for cell in row)
#check use for dublicate row data remove
        if row_cleaned not in check and all(cell is not None and str(cell).strip() != 'nan' for cell in row_cleaned):
           # print(row_cleaned)
            current_index = len(check) + 1
            check[current_index] = row_cleaned
            unique_row.append(list(row_cleaned))




def recursive_filter(datafrem, length, first_row, last_row, check, unique_row):
    # Base case
    if first_row >= last_row or first_row >= len(datafrem):
        return

    row_1 = datafrem.iloc[first_row]
    row_2 = datafrem.iloc[last_row-1]
    mid = int((first_row + last_row) // 2)


    for_clean (row_1,length,check,unique_row)
    for_clean (row_2,length,check,unique_row)

    # Recursive call
    recursive_filter(datafrem, length, first_row + 1, mid, check, unique_row)
    recursive_filter(datafrem, length, mid + 1, last_row , check, unique_row)





check = {}
unique_rows = []
length = len(data.iloc[0])  # column count

recursive_filter(df, length, 0, len(df)-1, check, unique_rows)



```
